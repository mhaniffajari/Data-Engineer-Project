# Anomaly Detection

## IQR

```python
Q1 = df['perc diff price'].quantile(0.25)
Q3 = df['perc diff price'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5*IQR
upper_bound = Q3 + 1.5*IQR
```

```python
df[(df['perc diff price']<lower_bound) | (df['perc diff price']>upper_bound)]
```

```python
df[(df['perc diff price']<lower_bound) | (df['perc diff price']>upper_bound)].groupby(['dealer_origin'])['total'].count().reset_index().sort_values('total',ascending=False)
```

```python
mean = df['perc diff price'].mean()
std = df['perc diff price'].std()
lower_bound = mean - 3*std
upper_bound = mean + 3*std
```

```python
df[(np.abs((df['perc diff price']-mean)/std)>3)]
```

```python
df[(np.abs((df['perc diff price']-mean)/std)>3)].groupby('dealer_origin')['total'].count().reset_index().sort_values('total',ascending=False)
```

```python
df_anomali = df[df['perc diff price']>=3].groupby(['motorcycle_name','motorcycle_group','dealer_origin'])['total'].count().reset_index().sort_values('total',ascending=False).rename({'total':'total_anomali'},axis=1)
```

```python
df_anomali
```

```python
df_normal = df.groupby(['motorcycle_name','motorcycle_group','dealer_origin'])['total'].count().reset_index().sort_values('total',ascending=False).rename({'total':'total_normal'},axis=1)
```

```python
df_normal
```

```python
df_merged = pd.merge(df_anomali, df_normal, on=['motorcycle_name', 'motorcycle_group', 'dealer_origin'], how='inner')

df_merged['perc anomali'] = df_merged['total_anomali']/df_merged['total_normal']*100
df_merged
```
